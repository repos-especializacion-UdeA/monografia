#### Segmentacion de la señal

Ahora vamos a proceder a realizar la segmentación de la señal.

```python
# Ejemplo de uso
w_size = 30  # Tamaño de la ventana
overlap = 0.33  # 50% de solapamiento
# ventanas = segment_signal(signal_emg_1_filt, window_size = w_size, overlap = overlap)
ventanas = segment_signal_as_matrix(signal_emg_1_filt.iloc[:,0], window_size = w_size, overlap = overlap)
#print(ventanas[0])

print(f"Se generaron {len(ventanas)} ventanas de tamaño 30 con 33% de solapamiento.")
print(f"Tamaño de la matriz: {ventanas.shape}")
```

Vamos a tranponer para analizar y aplicar medidas:


```python
ventanas_T = ventanas.T
print(ventanas_T.shape)
```

```python
columna_0 = ventanas_T[:, 0]
print(columna_0)
```

```python
# Calcular RMS para cada columna de ventanas_T
rms_v0 = calculate_rms(columna_0)
print(rms_v0)

```

```python
# Calcular RMS para todas las columnas de ventanas_T
rms_vector = np.sqrt(np.mean(ventanas_T**2, axis=0))
print(rms_vector)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rms_vector, label='RMS de las ventanas', color='blue')
plt.title('RMS de las Ventanas')
plt.xlabel('Índice de Ventana')
plt.ylabel('RMS')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rms_vector[:1000], label='RMS de las ventanas (Primeros 1000 puntos)', color='blue')
plt.title('RMS de las Ventanas (Primeros 1000 puntos)')
plt.xlabel('Índice de Ventana')
plt.ylabel('RMS')
plt.legend()
plt.grid(True)
plt.show()
```
